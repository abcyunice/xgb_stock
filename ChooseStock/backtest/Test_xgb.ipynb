# 引入随机数

```python
import xgboost as xgb
import pandas as pd
import numpy as np
import datetime

import pickle
```

```python
np.random.seed(1)

df_sh_train_X = np.random.normal(size=(20000,10))
df_sh_train_y = np.random.normal(size=(20000,1))

df_sh_test_X = np.random.normal(size=(1000,10))
df_sh_test_y = np.random.normal(size=(1000,1))

myxgb = xgb.XGBRegressor(n_estimators=10000, learning_rate=0.01, max_depth=15)
eval_set = [(df_sh_test_X, df_sh_test_y)]
myxgb.fit(df_sh_train_X, df_sh_train_y, eval_metric="rmse", eval_set=eval_set, verbose=False,
          early_stopping_rounds=100)

train_score = myxgb.score(df_sh_train_X, df_sh_train_y)
valid_score = myxgb.score(df_sh_test_X, df_sh_test_y)

print(train_score)
print(valid_score)
```
